```julia
using Plots, OhMyREPL

struct Line
    gridx::Vector{Float64}
    gridy::Vector{Float64}
end

struct Canvas
    lines::Vector{Line}
end

mutable struct Turtle
    x::Float64
    y::Float64
    direction::Float64
    pen::Bool
    canvas::Canvas
end

function Turtle() # Creates the Turtle
    plot()
    Turtle(0, 0, 90, false, Canvas([Line([], [])]))  # Faces the right while the pen is up
end

function Move!(t, i)
    if t.pen   # If the pen is down, start writing at the start
        append!(t.canvas.lines[end].gridx, t.x)
        append!(t.canvas.lines[end].gridy, t.y)
    end
    t.y += cosd(t.direction) * i  # Moves the Turtle in the y direction dependent on its angle
    t.x += sind(t.direction) * i  # Moves the Turtle in the x direction dependent on its angle
    if t.pen   # if the pen is down, stop writing at this location
        append!(t.canvas.lines[end].gridx, t.x)
        append!(t.canvas.lines[end].gridy, t.y)
    end
end
function Turn!(t, k)  # Adds the angle to the Turtles direction
    t.direction = mod((k + t.direction), 360)
end
function PenUp!(t)  # Takes the pen of the paper
    t.pen = false
end
function PenDown!(t) #Puts the pen down
    t.pen = true
    push!(t.canvas.lines, Line([], []))
end

function draw(t::Turtle)
    [plot!(line.gridx, line.gridy, color=:black, label=nothing) for line in t.canvas.lines] |> last
end

function test_turtle()
    t = Turtle()
    for i in 1:10
        PenDown!(t)
        Turn!(t, 270)
        Move!(t, 1)
        PenUp!(t)
        Turn!(t, 100)
        Move!(t, 1)
    end
    draw(t)
end

```

```julia
function a_turtle()
    t = Turtle()
    for i in 1:20
        PenDown!(t)
        Move!(t, 2)
        Turn!(t, 90)
        PenUp!(t)
        Move!(t, 1)
        Turn!(t, 90)
        PenDown!(t)
        Move!(t, 2)
        Turn!(t, 270)
        PenUp!(t)
        Move!(t, 1)
        Turn!(t, 270)
    end
    draw(t)
end

a_turtle()
```

```julia

function c_turtle()
    t = Turtle()
    PenDown!(t)
    k = 5
    for i in 1:k
        Turn!(t, 270)
        Move!(t, 1)
        Turn!(t, 90)
        Move!(t, 1)
    end
    draw(t)
end

c_turtle()
```

```julia


function d_turtle()
    t = Turtle()
    PenDown!(t)
    k = 50 # How Many Steps
    n = 30 # How Many Squares

    for i in 1:k   # Ladder
        Turn!(t, 270)
        Move!(t, 1)
        Turn!(t, 90)
        Move!(t, 1)
    end

    for i in 1:4 #  Square
        Move!(t, 2)
        Turn!(t, 270)
    end

    for i in 1:n  # Tower
        PenUp!(t)
        Turn!(t, 270)
        Move!(t, 2)
        PenDown!(t)
        for i in 1:3   #|_|  A Square with its bottom missing
            Move!(t, 2)
            Turn!(t, 90)
        end
        PenUp!(t) # Resets the turtle for next loop
        Move!(t, 2)
        Turn!(t, 180)
    end  # Ending triangle
    Turn!(t, 270)
    Move!(t, 2)
    PenDown!(t)
    Turn!(t, 30)
    Move!(t, 2)
    Turn!(t, 120)
    Move!(t, 2)
    draw(t)
end
d_turtle()
```

```julia
function pi_turtle()
    t = Turtle()
    PenDown!(t)
    Move!(t,7)
    Turn!(t,90)
    for i in 1:22
        Move!(t,1)
        Turn!(t,360/22)
    end
    draw(t)
end
pi_turtle()
```
